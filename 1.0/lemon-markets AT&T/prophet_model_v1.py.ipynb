```python
import pandas as pd
from prophet import Prophet

df = pd.read_csv('data.csv')

model = Prophet(interval_width=0.95, daily_seasonality=True, changepoint_prior_scale=1, seasonality_prior_scale=0.01, holidays_prior_scale=1)
#{'changepoint_prior_scale': 0.01, 'seasonality_prior_scale': 0.1, 'holidays_prior_scale': 0.01}
m = model.fit(df)


future = m.make_future_dataframe(periods=30)
forecast = m.predict(future)


print(forecast)
plot1 = m.plot(forecast)
```
